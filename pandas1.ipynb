Groupby

```python
d=pd.DataFrame({'Product':['BULB',"BULB","FAN","FAN","WASHINF MACH"],"Type":['A',"B",'A',"A",'B'],'EC':[450,300,500,450,900]})
d
```

```python
d.groupby(['Product']).sum()
```

```python
d.groupby(['Product']).mean()
```

```python
d.groupby(['Type']).mean()
```

```python
d.groupby([d.Product,d.Type]).sum()
```
