Hierarchical Indexing

```python
A=[['BULB',"BULB","BULB","Fan","Fan","WashingMach","WashingMach"],['A',"B","C","A","B","A","B"]]
idx=pd.MultiIndex.from_arrays(A,names=('ProductName',"Type"))
df1=pd.DataFrame({'EC':[56,78,43,26,78,89,77]},index=idx)
df1
```

```python
df1.groupby(level=0).sum()
```

**Where Function**

dataFrame.where(cond,action)

```python
df3=pd.DataFrame(np.arange(10).reshape(5,2),columns=['A','B'])
df3
```

```python
df3.where(df3<5,-df3)
```

```python
df3.where((df3<5) | (df3%3),-df3)
```

**dataFrame.merge(df1,df2)**

```python
df1=pd.DataFrame({'E':['B',"J",'L','S'],
                  'G':['A','E','E','H']})
df2=pd.DataFrame({'E':['L','B','J','S'],
                  "H":[2004,2008,2012,2016]})
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df4=pd.DataFrame({'G':['A',"E","H"],'S':['C','G','S']})
df4
```

```python
pd.merge(df3,df4,on='G')
```

dataFrame.pivot_table(....)

```python
from seaborn import load_dataset
data=sns.load_dataset('titanic')
data.head()
```

```python
data.rename(columns={'sex':'gender'},inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(['deck'],axis=1,inplace=True)
data.head()
```

```python
data.groupby(['gender','class'])['survived'].mean()
```

```python
data.groupby(['gender','class'])['survived'].mean().unstack()
```

```python
data.pivot_table('survived',index='gender',columns='class',aggfunc='mean')
```

```python
agePartition=pd.cut(data['age'],[0,18,80])
agePartition
```

```python
data.pivot_table('survived',['gender',agePartition],'class')
```

```python
farePartition=pd.cut(data['fare'],2)
farePartition
```

```python
data.pivot_table('survived',['gender',agePartition],[farePartition,'class'])
```

```python
data.pivot_table(index='gender',columns='class',aggfunc={'survived':'sum','fare':'mean'})
```
